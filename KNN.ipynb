```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('GSE71766.csv', index_col=0)
df.head()
```

```python
df_label = df[['stage']]
df.drop(columns=['stage'], inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df_label
```

```python
df_label.value_counts()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df_label, x='stage')
plt.title('Distribution of Labels')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, df_label, test_size=0.2, random_state=0, stratify = df_label)
```

```python
print("X_train:", X_train.shape, "X_test", X_test.shape)
print("y_train:", y_train.shape, "y_test", y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# y_train'i sÃ¼tun vektÃ¶rÃ¼ne dÃ¶nÃ¼ÅŸtÃ¼rme
y_train_reshaped = y_train.values.ravel()

# KNN sÄ±nÄ±flandÄ±rÄ±cÄ±yÄ± oluÅŸturma
clf = KNeighborsClassifier()

# EÄŸitim setlerini kullanarak modeli eÄŸitme
clf.fit(X_train, y_train_reshaped)

# Test veri kÃ¼mesi iÃ§in yanÄ±tÄ± tahmin etme
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Macro Precision:",metrics.precision_score(y_test, y_pred,average='macro'))
print("Macro Recall:",metrics.recall_score(y_test, y_pred,average='macro'))
```

```python
target_names = ['Uninfected', 'Influenza virus infected', 'Rhino virus infected','Influenza & Rhino virus infected' ]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
# DataFrame'in kopyasÄ±nÄ± al
df_label_copy = df_label.copy()

# DeÄŸerleri gÃ¼ncelle
df_label_copy.loc[df_label_copy["stage"] == 1, "stage"] = 1 # uninfected
df_label_copy.loc[df_label_copy["stage"].isin([2, 3, 4]), "stage"] = 2 # infected

# SonuÃ§larÄ± kontrol et
df_label_copy["stage"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# Veri setini eÄŸitim ve test setlerine ayÄ±rma
X_train, X_test, y_train, y_test = train_test_split(df, df_label_copy, test_size=0.2, random_state=0, stratify=df_label_copy)

# y_train'i sÃ¼tun vektÃ¶rÃ¼ne dÃ¶nÃ¼ÅŸtÃ¼rme
y_train_reshaped = y_train.values.ravel()

# KNN sÄ±nÄ±flandÄ±rÄ±cÄ±yÄ± oluÅŸturma
clf = KNeighborsClassifier()

# EÄŸitim setlerini kullanarak modeli eÄŸitme
clf.fit(X_train, y_train_reshaped)

# Test veri kÃ¼mesi iÃ§in tahminde bulunma
y_pred = clf.predict(X_test)

# DeÄŸerlendirme
print(classification_report(y_test, y_pred, target_names=['Uninfected', 'Infected']))
```

```python
from sklearn.model_selection import cross_val_score

# y_train'i sÃ¼tun vektÃ¶rÃ¼ne dÃ¶nÃ¼ÅŸtÃ¼rme
y_train_reshaped = y_train.values.ravel()

scores = cross_val_score(clf, X_train, y_train_reshaped, cv=10)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.model_selection import GridSearchCV
# y_train'i sÃ¼tun vektÃ¶rÃ¼ne dÃ¶nÃ¼ÅŸtÃ¼rme
y_train_reshaped = y_train.values.ravel()

# Hyperparameter Optimization
parameters = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}

knn = KNeighborsClassifier()

# Grid search'Ã¼ Ã§alÄ±ÅŸtÄ±rma
grid_obj = GridSearchCV(knn, parameters, cv=5)
grid_obj = grid_obj.fit(X_train, y_train_reshaped)

# En iyi parametre yapÄ±landÄ±rmasÄ±nÄ± gÃ¶sterme
best_parameters = grid_obj.best_params_
print(best_parameters)

# En iyi parametre kombinasyonunu belirlemek iÃ§in clf'yi ayarlama
best_clf = grid_obj.best_estimator_

# Test veri kÃ¼mesi iÃ§in tahminde bulunma
y_pred = best_clf.predict(X_test)

# SÄ±nÄ±flandÄ±rma raporunu yazdÄ±rma
print(classification_report(y_test, y_pred, target_names=['Uninfected', 'Infected']))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=best_clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_clf.classes_)
disp.plot()
plt.show()
```
